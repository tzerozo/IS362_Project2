To tidy up "Unemployment Data" 

```python
import pandas as pan
import numpy as num
```

NOTE: the data file is alreay on github as raw. 
TASK1: Read the .csv file into Python

```python
data1 = pan.read_csv("https://raw.githubusercontent.com/amberferger/DATA607_Project2/master/Unemployment%20Data.csv")
```

NOTE: check the data

```python
data1.head(12)
```

NOTE: 
1. Row 1 is empty, we could remove it.
2. Col 1 is "Unnamed", we could rename it to "YEAR"
3. Col 1 is float, we should change it to "int".
4. There are many missing data points aka NaN, will drop/remove all columns with NaN values.

TASK2.1: Remove Row 1

```python
data1 = data1.drop(0)
```

NOTE: Check to confirm

```python
data1.head(12)
```

TASK2.2: Change Col 1 name to appropriate name "YEAR"

```python
data1.rename (columns = {'Unnamed: 0': 'YEAR'}, inplace=True)
```

TASK2.3: Change Col 1 data type into int (1990.0 into 1990) using astype(int) since Col 1 is already numeric. 

```python
data1["YEAR"] = data1["YEAR"].astype(int)
```

```python
NOTE: Check to confirm
```

```python
data1.head(12)
```

TEST: Checking to see if there are any rows without "NaN value".

```python
data1.dropna()
```

TASK2.4: Dropping columns having any NaN values

```python
data1 =data1.dropna(axis=1)
```

```python
data1.head(30)
```

NOTE: After dropping "NaN" value contained Columns, check how many columns left.

```python
col = data1.columns
print(len(col))
```
